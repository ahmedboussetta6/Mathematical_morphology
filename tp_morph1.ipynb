```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
imgset=np.zeros((15,15))
imgset[2:13,2:6]=1
imgset[2:13,7:10]=1
imgset[5:13,6]=1
imgset[7,10:14]=1
imgset[6:7,12]=1

plt.imshow(imgset,cmap='gray')
```

```python
#disque
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3,3))
erosion = cv2.erode(imgset, kernel, iterations=1)
dilatation = cv2.dilate(imgset, kernel, iterations=1)

plt.figure(figsize=(15,15))

plt.subplot(1,3,2)
plt.imshow(erosion, cmap='gray', interpolation='nearest')
plt.title("Érosion")

plt.figure(figsize=(15,15))
plt.subplot(1,3,3)
plt.imshow(dilatation, cmap='gray', interpolation='nearest')
plt.title("Dilatation")
```

```python

img = cv2.imread("gray.JPG", cv2.IMREAD_GRAYSCALE)

kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (30,30))

erosion = cv2.erode(img, kernel, iterations=1)
dilatation = cv2.dilate(img, kernel, iterations=1)


complement_original = cv2.bitwise_not(img)
complement_erosion = cv2.bitwise_not(erosion)
complement_dilatation = cv2.bitwise_not(dilatation)

plt.figure(figsize=(20, 15))

plt.subplot(2, 3, 4)
plt.imshow(complement_original, cmap='gray')
plt.title("Complément - Originale")

plt.subplot(2, 3, 5)
plt.imshow(complement_erosion, cmap='gray')
plt.title("Complément - Érosion")

plt.subplot(2, 3, 6)
plt.imshow(complement_dilatation, cmap='gray')
plt.title("Complément - Dilatation")

plt.tight_layout()
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread("gray.JPG", cv2.IMREAD_GRAYSCALE)

kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (30,30))

erosion = cv2.erode(img, kernel, iterations=1)
dilatation = cv2.dilate(img, kernel, iterations=1)

ouverture = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)

fermeture = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)

complement_original = cv2.bitwise_not(img)
complement_erosion = cv2.bitwise_not(erosion)
complement_dilatation = cv2.bitwise_not(dilatation)
complement_ouverture = cv2.bitwise_not(ouverture)
complement_fermeture = cv2.bitwise_not(fermeture)

plt.figure(figsize=(20, 12))

plt.subplot(2, 5, 9)
plt.imshow(complement_ouverture, cmap='gray')
plt.title("Complément - Ouverture")

plt.subplot(2, 5, 10)
plt.imshow(complement_fermeture, cmap='gray')
plt.title("Complément - Fermeture")

plt.tight_layout()
plt.show()
```
