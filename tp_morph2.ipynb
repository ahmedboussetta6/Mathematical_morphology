```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
imgset=np.zeros((15,15))
imgset[2:13,2:6]=1
imgset[2:13,7:10]=1
imgset[5:13,6]=1
imgset[7,10:14]=1
imgset[6:7,12]=1

plt.imshow(imgset,cmap='gray')
```

```python
import cv2
import matplotlib.pyplot as plt

kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3,3))


ouverture = cv2.morphologyEx(imgset, cv2.MORPH_OPEN, kernel)
fermeture = cv2.morphologyEx(imgset, cv2.MORPH_CLOSE, kernel)

# Affichage
plt.figure(figsize=(15, 8))

plt.subplot(1, 4, 3)
plt.imshow(ouverture, cmap='gray')
plt.title("Ouverture")

plt.subplot(1, 4, 4)
plt.imshow(fermeture, cmap='gray')
plt.title("Fermeture")

plt.tight_layout()
plt.show()
```

```python

img = cv2.imread('text.png', cv2.IMREAD_GRAYSCALE)

kernel1 = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))

ouverture = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel1)

fermeture = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel1)

plt.figure(figsize=(15, 6))

plt.subplot(1, 3, 1)
plt.imshow(img, cmap='gray')
plt.title("Image originale")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(ouverture, cmap='gray')
plt.title("Ouverture")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(fermeture, cmap='gray')
plt.title("Fermeture")
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
brain=cv2.imread('brain.png', cv2.IMREAD_GRAYSCALE)

_, img_binaire = cv2.threshold(brain, 127, 255, cv2.THRESH_BINARY)

kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (4,4))

ouverture = cv2.morphologyEx(img_binaire, cv2.MORPH_OPEN, kernel)

fermeture = cv2.morphologyEx(img_binaire, cv2.MORPH_CLOSE, kernel)


plt.figure(figsize=(15, 6))

plt.subplot(1, 4, 2)
plt.imshow(img_binaire, cmap='gray')
plt.title("Image binaire")
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(ouverture, cmap='gray')
plt.title("Ouverture")
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(fermeture, cmap='gray')
plt.title("Fermeture")
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
img_color = cv2.imread('melanome.jpg')

img_gray = cv2.cvtColor(img_color, cv2.COLOR_BGR2GRAY)

kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (50,50))

erosion = cv2.erode(img_gray, kernel, iterations=1)

dilatation = cv2.dilate(img_gray, kernel, iterations=1)

plt.figure(figsize=(15, 6))

plt.subplot(1, 3, 1)
plt.imshow(img_gray, cmap='gray')
plt.title("Image en niveaux de gris")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(erosion, cmap='gray')
plt.title("rosion")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(dilatation, cmap='gray')
plt.title("Dilatation")
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
img_gray = cv2.cvtColor(img_color, cv2.COLOR_BGR2GRAY)

kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (20,20))

ouverture = cv2.morphologyEx(img_gray, cv2.MORPH_OPEN, kernel)

fermeture = cv2.morphologyEx(img_gray, cv2.MORPH_CLOSE, kernel)


plt.figure(figsize=(15, 6))

plt.subplot(1, 4, 2)
plt.imshow(img_gray, cmap='gray')
plt.title("Image binaire")
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(ouverture, cmap='gray')
plt.title("Ouverture")
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(fermeture, cmap='gray')
plt.title("Fermeture")
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python


kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3,3))

erosion = cv2.erode(img_gray, kernel, iterations=1)
dilatation = cv2.dilate(img_gray, kernel, iterations=1)

gradient_exterieur = cv2.subtract(dilatation, img_gray)

gradient_interieur = cv2.subtract(img_gray, erosion)

gradient_morphologique = cv2.subtract(dilatation, erosion)

laplacian=cv2.Laplacian(img_gray,cv2.CV_16S,ksize=3)

laplacian=cv2.convertScaleAbs(laplacian)

plt.figure(figsize=(15, 10))

plt.subplot(2, 3, 1)
plt.imshow(img_gray, cmap='gray')
plt.title("Image originale")
plt.axis('off')

plt.subplot(2, 3, 2)
plt.imshow(gradient_interieur, cmap='gray')
plt.title("Gradient intrieur")
plt.axis('off')

plt.subplot(2, 3, 3)
plt.imshow(gradient_exterieur, cmap='gray')
plt.title("Gradient extrieur")
plt.axis('off')

plt.subplot(2, 3, 4)
plt.imshow(gradient_morphologique, cmap='gray')
plt.title("Gradient morphologique total")
plt.axis('off')

plt.subplot(2, 3, 5)
plt.imshow(laplacian, cmap='gray')
plt.title("Laplacien morphologique")
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
tumeur = cv2.imread('RMS_516_821_fig01_i1200.jpg', cv2.IMREAD_GRAYSCALE)

kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3,3))

erosion = cv2.erode(tumeur, kernel, iterations=1)
dilatation = cv2.dilate(tumeur, kernel, iterations=1)

gradient_exterieur = cv2.subtract(dilatation, tumeur)

gradient_interieur = cv2.subtract(tumeur, erosion)

gradient_morphologique = cv2.subtract(dilatation, erosion)

laplacian=cv2.Laplacian(tumeur,cv2.CV_16S,ksize=3)

laplacian=cv2.convertScaleAbs(laplacian)

plt.figure(figsize=(15, 10))

plt.subplot(2, 3, 1)
plt.imshow(tumeur, cmap='gray')
plt.title("Image originale")
plt.axis('off')

plt.subplot(2, 3, 2)
plt.imshow(gradient_interieur, cmap='gray')
plt.title("Gradient intrieur")
plt.axis('off')

plt.subplot(2, 3, 3)
plt.imshow(gradient_exterieur, cmap='gray')
plt.title("Gradient extrieur")
plt.axis('off')

plt.subplot(2, 3, 4)
plt.imshow(gradient_morphologique, cmap='gray')
plt.title("Gradient morphologique total")
plt.axis('off')

plt.subplot(2, 3, 5)
plt.imshow(laplacian, cmap='gray')
plt.title("Laplacien morphologique")
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (29,29))


top_hat = cv2.morphologyEx(tumeur, cv2.MORPH_TOPHAT, kernel)

black_hat = cv2.morphologyEx(tumeur, cv2.MORPH_BLACKHAT, kernel)

plt.figure(figsize=(15, 6))

plt.subplot(1, 3, 1)
plt.imshow(tumeur, cmap='gray')
plt.title("Image originale")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(top_hat, cmap='gray')
plt.title("Top-hat ")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(black_hat, cmap='gray')
plt.title("Top-hat conjugu ")
plt.axis('off')

plt.tight_layout()
plt.show()
```
